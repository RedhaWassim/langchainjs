# ChatEdenAI

EdenAI is a platform that aggregates and provides access to multiple Large Language Models (LLMs) under a single account. With EdenAI, users can interact with various LLMs from different providers using a unified API. This eliminates the need for managing multiple accounts .

more informations : [EdenAI](https://edenai.co/)

You can find all the models we support here: [EdenAI Featured Models](https://docs.edenai.co/reference/text_chat_create)

## Setup

To access EdenAI models you'll need to create an Edenai account and get an API key at https://www.edenai.co

### Credentials

Head to [this page](https://app.edenai.run/admin/api-settings) to sign up to EdenAI and generate an copy your api key. Once you've done this set the EDENAI_API_KEY environment variable:

```bash
export EDENAI_API_KEY="your-api-key"
```
